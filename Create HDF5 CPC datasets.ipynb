```python
import h5py
import numpy as np
import  simplejson as json,os
from tqdm import tqdm
```

# For Training set

```python
dataset_name='CPC_5_head_sample_per_subgroup'
PATH='/data/cf_dataset/CPC/US_CPC_2010_2019_subgroup_200/bert-cpc'
#dataset_file = os.path.join(PATH,'head.json')
dataset_file = '/data/cf_dataset/CPC/US_CPC_2010_2019_subgroup_200/CPC_5_head_sample_per_subgroup.json'
dataset_hdf5_file = os.path.join(PATH,'uspto_cpc' + '.hdf5')
```

```python
#Pre-process data by removing json that's unable to process

def read_dataset(filename,size=10000):
    count =0
    with open(filename,'r') as f:
        for line in tqdm(f,'read_dataset',total=size): #575020
            try:
                json_text = json.dumps(json.loads(line))
                count+=1
                yield json_text
                if count >=size:
                    return
            except:
                pass
        
```

```python
print(dataset_hdf5_file)
print(dataset_name)
```

```python
#Writing
with h5py.File(dataset_hdf5_file, "w") as f:
    max_length = 608717

    dt = h5py.special_dtype(vlen=unicode)

    dset = f.create_dataset(dataset_name,(max_length,), dtype=dt)
    #print(dset.name)
    #print(f.keys())
    #set values
    idx = 0
    for sample in read_dataset(dataset_file,size=max_length):
        dset[idx]= np.str(sample)
        #print(idx)

        idx+=1

    #dset[...] = [np.str(u'col' +str(i)) for i in range(max_length)]
    #print(dset[:1])
```

```python
#Reading
f  = h5py.File(dataset_hdf5_file, "r")
dset = f[dataset_name]
print(dset.size)
print(json.loads(dset[544100])['pubId'])

f.close()
```
